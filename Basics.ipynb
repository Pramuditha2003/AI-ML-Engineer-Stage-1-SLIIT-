```python
import matplotlib.pyplot as plt

import numpy as np
import math
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig,axes =  plt.subplots(2,3)

import numpy as np
x = np.arange(1,5,0.2)
axes[0][0].plot(x,x*x)
axes[0][0].set_title('square')
axes[0][1].plot(x,np.sqrt(x))
axes[0][1].set_title('square root')
axes[0][2].plot(x,np.exp(x))
axes[0][2].set_title('exp')
axes[1][0].plot(x,np.log10(x))
axes[1][0].set_title('log')
axes[1][1].plot(x, 2*x + 5)
axes[1][1].set_title("Linear")
axes[1][2].plot(x, 2*x**2 - 11*x + 10)
axes[1][2].set_title("Polynomial")
plt.tight_layout()

plt.show()
```
